### calculating peak hours of sales

```python
pizza_merge["time"]= pd.to_datetime(pizza_merge["time"], format= "%H:%M:%S").dt.time #first convert to datetime datatype
```

```python
pizza_merge.groupby([x.hour for x in pizza_merge["time"]])["price"].sum().sort_values(ascending=False).index[0] #calling the hour function... from the result, more purchases are within lunch time
```

### calculating total sales made each day of the week

```python
pizza_merge["date"]= pd.to_datetime(pizza_merge["date"]) # change the date column to datetime datatype
```

```python
pizza_merge["day_of_the_week"]=pizza_merge["date"].dt.day_name() #calling the day function
```

```python
pizza_merge.groupby("day_of_the_week")["price"].sum().sort_values(ascending=False) #suming up daily sales and arranging according to the highest sales per day
```

```python
pizza_merge.groupby('pizza_type_id')["quantity"].sum().sort_values(ascending=False).head(5) #grouping by top 5 best selling pizza
```

```python
pizza_merge["month"]=pizza_merge["date"].dt.month_name() #calling the month function
```

```python
pizza_merge.groupby("month")["price"].sum().sort_values(ascending=False) #grouping by sales made in each month from the highest to the lowest
```

```python
pizza_merge.groupby("pizza_type_id")["quantity"].sum().sort_values(ascending=True).head(10) #pizza types performing poorly ranked from the lowest
```
